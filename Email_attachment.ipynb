```python
import smtplib
import os

from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from email.mime.base import MIMEBase
from email import encoders 

#Setup SMTP port and server name

smtp_port = 587                 #Standard secure SMTP port 
smtp_server = "smtp.gmail.com"  #Google SMTP Server

email_from = "sl2ecde55@gmail.com"
email_list = ["20bec081@nirmauni.ac.in"]
pswd = "xasaxcnmgxchsrqb"
subject = "Result Announcement!" 

FILELIST = os.listdir("./SL_assignment/") #Enter into that folder whose files are to be sent as attachments
#To access each file in FILELIST use a for loop
def send_mails(email_list):
    for filename in FILELIST:
        for person in email_list:
            #Made the body of email 
            body = f"""
            Greetings of the Day!!!

            The Organization appreciates all the participants 
            for actively participating in the event. We extend
            our heartfelt Congratulations to all the winners. 
            We look forward to meeting you again at our events.            
            
            Winners can find their certificates in the attachment.  
            """
            #make a MIME object to define parts of the email
            msg = MIMEMultipart()
            msg['From'] = email_from
            msg['To'] = person
            msg['Subject'] = subject

            #Attach the body of the message 
            msg.attach(MIMEText(body, 'plain'))
                                       
            #Open the file in python as binary form
            attachment = open(filename, 'rb') #r means read and b means binary

            #Encode as base 64
            attachment_package = MIMEBase('application', 'octet-stream')  #Attachment package as application/octet-stream   
            attachment_package.set_payload((attachment).read())  #We are loading different bits so that the attachment is downloadable
            encoders.encode_base64(attachment_package)  #Encode file in ASCII characters to send by email
            attachment_package.add_header('Content-Disposition',"attachment; filename= " + filename)  #We are adding some headers to attachment part
            msg.attach(attachment_package)   # Lastly we are attaching to the msg object

            #Convert message to string
            text = msg.as_string()  # This line will fire the email 

            #Connect with the server 
            print("Connecting to server...")
            server = smtplib.SMTP(smtp_server, smtp_port)
            server.starttls()
            server.login(email_from, pswd)
            print("Successfully connected to the Server :)")
            
            #Send emails to person
            print(f"Sending email to: {person} ...")
            server.sendmail(email_from, person, text)
            print(f"Email sent to: {person}")
            server.quit()
            
send_mails(email_list)
```
